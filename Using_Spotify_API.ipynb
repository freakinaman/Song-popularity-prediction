```python
# creating datafram to store these values
import pandas as pd

track_feature = pd.DataFrame(sp.audio_features(track_uris))
track_feature.head(100)
```

```python
track_uri = []
track_name = []
artist_info = []
artist_name = []
artist_pop = []
artist_genres = []
track_pop = []
album = []

for track in sp.playlist_tracks(playlist_URI)["items"]:
  #URI
  track_uri.append(track["track"]["uri"])

  #track name
  track_name.append(track["track"]["name"])

  #Main artist
  artist_uri = track["track"]["artists"][0]["uri"]
  ar_info = sp.artist(artist_uri)
  artist_info.append(ar_info)

  #name popularity and genre
  artist_name.append(track["track"]["artists"][0]["name"])
  artist_pop.append(ar_info["popularity"])
  artist_genres.append(ar_info["genres"])

  #album
  album.append(track["track"]["album"]["name"])

  #popularity of the track
  track_pop.append(track["track"]["popularity"])
```

```python
extra_data = pd.DataFrame(list(zip(track_uri, track_name, artist_info, artist_name, artist_pop,
                                   artist_genres, track_pop, album)),
                          columns =['Track identifier', ' Track name', 'Artist information', 'Artist name',
                                    'Artist Pop', 'Genre of Artist', 'Track Pop', 'Album name'])
```

```python
# merging the dataframe to get the whole data

finalData = pd.concat([track_feature, extra_data], axis=1)
finalData.head()
```
